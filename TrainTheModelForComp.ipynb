```python
df=pd.DataFrame()
#print(IPathFile)
#df=pd.read_csv(Index_File,index_col=None)
df=pd.read_csv(Index_File,index_col=None)
df=df.drop(columns=['Unnamed: 0','Index','ID','Date'])

X=df.iloc[:,df.columns!='label']
y=df.label
df.head()
```

```python
#SPLITTING DATA INTO TESTING AND TRAINING
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
X_val, X_final, y_val, y_final = train_test_split(X_test, y_test,test_size=0.50, random_state=42)
```

```python

#Path='/dsa/home/pg3fy/jupyter/IR/COPD/NEW_WORK_For_IE_Ratio_AND_FeatureExtraction_AND_Clustering_PythonPeakDetection/PlottingCombined/'
#Index_File=IPath+'Indexes.csv'






#PARAMS TO BE TESTED IN GRID
param_grid = {
    'bootstrap': [True],
    'max_depth': [5,10, 20, 30, 40],
    'n_estimators': [25,50,70,100, 130, 150, 200]
}
```

```python
#mODEL Training
Model=RandomForestClassifier()#,class_weight=class_weight)
#scores = cross_val_score(clf, X_train,y_train, cv=5)
clf2 = GridSearchCV(estimator=Model, cv=10, n_jobs=4,param_grid=param_grid,verbose=2)
clf2.fit(X_train, y_train)
print("Training Score: ",clf2.best_score_)
print(clf2.best_estimator_)

print("Statistics for testing the model with test data")
y_pred=clf2.predict(X_val)
print("Testing Score: ",accuracy_score(y_val,y_pred))
print("Confusion Matrix:===> ","\n", confusion_matrix(y_val, y_pred),"\n")
print("Classification Report:===> ","\n", classification_report(y_val, y_pred))
print("Best Accuracy Score:===> ", clf2.best_score_)

print("Labels in labelled data: ",sorted(Counter(df['label'].to_list()).items()))
#joblib.dump(clf2, 'RandomForest_birch_9.pkl')
    
```
